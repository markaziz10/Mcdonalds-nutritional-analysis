```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
sns.set_style("whitegrid")
```

```python
#read in the data
df = pd.read_csv('menu.csv')
```

```python
#check the head of the data
df.head()
```

```python
#check the shape of the df
df.shape
```

```python
#check the data types
df.dtypes
```

```python
#check if there are empty values
df.isnull().sum()
```

```python
##Exploratory Data Analysis
```

```python
plt.scatter(df['Category'], df['Calories'], edgecolors='r')
plt.xlabel('Category')
plt.ylabel('Calories')
plt.xticks(rotation=90)
plt.title('Category against Calories')
plt.figure(figsize=(30,15))
plt.show()
```

```python
sns.barplot(df['Category'], df['Total Fat'])
plt.xlabel('Category')
plt.ylabel('Total Fat')
plt.xticks(rotation=90)
plt.title('Category against Total Fat')
plt.figure(figsize=(30,15))
plt.show()
```

```python
sns.barplot(df['Category'], df['Sugars'])
plt.xlabel('Category')
plt.ylabel('Sugars')
plt.xticks(rotation=90)
plt.title('Category against Sugars')
plt.figure(figsize=(30,15))
plt.show()
```

```python
sns.barplot(df['Category'], df['Sodium (% Daily Value)'])
plt.xlabel('Category')
plt.ylabel('Sodium (%Daily Value)')
plt.xticks(rotation=90)
plt.title('Category against Sodium (% Daily Value)')
plt.figure(figsize=(30,15))
plt.show()
```

```python
sns.boxenplot(df['Category'], df['Protein'])
plt.xlabel('Category')
plt.ylabel('Protein')
plt.xticks(rotation=90)
plt.title('Category against Protein')
plt.figure(figsize=(30,15))
plt.show()
```

```python
sns.heatmap(df.corr(),cmap='inferno')
plt.title('df.corr()')
```

```python
sns.violinplot(x="Category", y="Calories", data=df,palette='rainbow')
```

```python
sns.set(rc={'figure.figsize':(23,8)})
sns.barplot(x='Category',y='Calories',data=df)
```

```python
calories = df.sort_values(by='Calories', ascending=False)[:10]
```

```python
sns.set(rc={'figure.figsize':(23,8)})
sns.barplot(x='Item',y='Calories',data=calories)
plt.xticks(rotation=90)
```
